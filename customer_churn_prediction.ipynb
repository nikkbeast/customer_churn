```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('customer_churn.csv')
```

```python
df.sample(5)
```

```python
df.drop(columns='customerID',axis=1,inplace=True)
```

```python
df.dtypes
```

```python
df.TotalCharges.values
```

```python
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
df1=df[df.TotalCharges!=' ']
```

```python
df1[pd.to_numeric(df1.TotalCharges,errors='coerce').isnull()]
```

```python
df1.TotalCharges=pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtypes
```

```python
tenure_churn_no=df1[df1.Churn=='No'].tenure
tenure_churn_yes=df1[df1.Churn=='Yes'].tenure
```

```python
plt.hist([tenure_churn_yes,tenure_churn_no],color=['green','red'],label=['churn=Yes','churn=NO'])
plt.xlabel('Tenure in months')
plt.ylabel('NO Of Customers')
plt.legend()
```

```python
mc_churn_no=df1[df1.Churn=='No'].MonthlyCharges
mc_churn_yes=df1[df1.Churn=='Yes'].MonthlyCharges
plt.hist([mc_churn_yes,mc_churn_no],color=['green','red'],label=['churn=Yes','churn=NO'])
plt.xlabel('Monthly Charges')
plt.ylabel('NO Of Customers')
plt.legend()
```

```python
def print_unique_col_value(df):
    for c in df:
        if df[c].dtypes=='object':
            print(f'{c}:{df[c].unique()}')
print_unique_col_value(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
df1
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df3 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'],dtype='int64')
df3.columns
```

```python
df3.sample(5)
```

```python
df3.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df3[cols_to_scale] = scaler.fit_transform(df3[cols_to_scale])
```

```python
for col in df3:
    print(f'{col}: {df3[col].unique()}')
```

```python
X = df3.drop('Churn',axis='columns')
y = df3['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
log.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nbgc=GaussianNB()
nbgc.fit(X_train,y_train)
nbgc.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
svc_rbf=SVC()
svc_rbf.fit(X_train,y_train)
svc_rbf.score(X_test,y_test)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
